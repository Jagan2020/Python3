### How to compare two images using histogram?

```python
from PIL import Image,ImageChops

img1 = Image.open('/home/mana/Pictures/chitra/sat1.jpg').histogram()
img2 = Image.open('/home/mana/Pictures/chitra/sat2.jpg').histogram()

if img1 == img2:
    print('Both images are same.')
else:
    print('Both images are not same.')
```

### How to identify  and save the changed location between two images?

```python
from PIL import Image,ImageChops
 

img1 = Image.open('/home/mana/Pictures/chitra/sat1.jpg')
img2 = Image.open('/home/mana/Pictures/chitra/sat2.jpg')
diff = ImageChops.difference(img1, img2)
 
if diff.getbbox():
        diff.save('/home/mana/Pictures/chitra/change-location.jpg')
```
