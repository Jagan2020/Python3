#### **Channel Methods**

```python
from paramiko import Channel

methods = [i for i  in dir(Channel) if not i.startswith('_')]
fmt = '{:25s}'*4

for fn in zip(*[iter(methods)] *4):
    print(fmt.format(*fn))
```

```python
from paramiko import Channel

methods = [i for i  in dir(Channel) if not i.startswith('_')]

for i in methods:
    print(i)
    print(getattr(Channel,i).__doc__)
    print('*'*50)
```
