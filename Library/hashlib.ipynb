#### **The md5sum is designed to verify data integrity using MD5 (Message Digest Algorithm 5).**

```python
from pathlib import Path
from hashlib import md5

p = Path('/home/mana/Work/test.csv').read_bytes()
print(md5(p).hexdigest())
```
